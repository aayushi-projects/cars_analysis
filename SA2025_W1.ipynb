### 2. How many cars have mpg  35?

```python
import pandas as pd

# Load the CSV again
df = pd.read_csv('Cars.csv')

# Ensure 'mpg' is numeric
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

# Filter and count cars with mpg >= 35
high_mpg_count = df[df['mpg'] >= 35].shape[0]

print("Number of cars with mpg  35:", high_mpg_count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
import pandas as pd

# Load data
df = pd.read_csv('Cars.csv')

# Ensure numeric types
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')

# Filter based on conditions
filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Find the most common origin
most_common_origin = filtered_df['origin'].mode()[0]

print("Most common origin for cars with horsepower > 100 and weight < 3000:", most_common_origin)

```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Cars.csv')

# Filter for cars from Japan
japan_cars = df[df['origin'] == 'japan']

# Ensure 'acceleration' is numeric
japan_cars['acceleration'] = pd.to_numeric(japan_cars['acceleration'], errors='coerce')

# Calculate and round the mean
mean_acceleration = round(japan_cars['acceleration'].mean(), 2)

print("Mean acceleration of cars from Japan:", mean_acceleration)

```

### 5. Which year had the highest average mpg?

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Cars.csv')

# Clean column names: strip spaces and convert to lowercase
df.columns = df.columns.str.strip().str.lower()

print("Columns:", df.columns)

model_year_col = [col for col in df.columns if 'year' in col][0]  # Find the correct year-related column

df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

avg_mpg_by_year = df.groupby(model_year_col)['mpg'].mean()

year_with_max_avg_mpg = avg_mpg_by_year.idxmax()
highest_avg = round(avg_mpg_by_year.max(), 2)


print(f"Year with highest average mpg: {year_with_max_avg_mpg} (Avg MPG: {highest_avg})")


```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Cars.csv")

# Convert 'horsepower' to numeric, coercing errors to NaN
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Drop rows with missing values in 'horsepower' or 'mpg'
df_clean = df.dropna(subset=['horsepower', 'mpg'])

# Compute the median mpg
median_mpg = df_clean['mpg'].median()

# Filter cars with mpg above the median
df_filtered = df_clean[df_clean['mpg'] > median_mpg].copy()

# Compute horsepower-to-weight ratio
df_filtered['hp_to_weight'] = df_filtered['horsepower'] / df_filtered['weight']

# Find the maximum horsepower-to-weight ratio
max_ratio = df_filtered['hp_to_weight'].max()

# Get the car(s) with the best ratio
best_cars = df_filtered[df_filtered['hp_to_weight'] == max_ratio]

# Display the relevant columns
print(best_cars[['name', 'mpg', 'horsepower', 'weight', 'hp_to_weight']])

```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Cars.csv")

# Convert 'horsepower' and 'model_year' to numeric
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['model_year'] = pd.to_numeric(df['model_year'], errors='coerce')

# Drop rows with missing mpg, horsepower, or model_year
df_clean = df.dropna(subset=['mpg', 'horsepower', 'model_year'])

# Group by model_year and origin to compute average mpg
avg_mpg_per_year_origin = df_clean.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Set the seaborn style
sns.set(style="whitegrid")

# Create the multi-line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg_per_year_origin, x='model_year', y='mpg', hue='origin', marker='o')

# Customize plot
plt.title('Evolution of Average MPG Over the Years by Origin', fontsize=16)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.legend(title='Origin')
plt.tight_layout()

# Show the plot
plt.show()

```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Cars.csv")

# Convert relevant columns to numeric
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

# Drop rows with missing values
df_clean = df.dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])

# Define hue order
hue_order = ['japan', 'europe', 'usa']

# Create scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_clean,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=hue_order,
    sizes=(20, 200),
    alpha=0.7,
    palette='muted'
)

# Add titles and labels
plt.title('Horsepower vs. Weight of Cars\nColored by Origin, Sized by MPG', fontsize=16)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight (lbs)', fontsize=12)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Cars.csv")

# Convert necessary columns to appropriate types
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')
df['model_year'] = pd.to_numeric(df['model_year'], errors='coerce')

# Drop rows with missing values in key columns
df_clean = df.dropna(subset=['mpg', 'model_year', 'name'])

# Group by car name and filter those that appear in multiple years
grouped = df_clean.groupby('name')

# Create a DataFrame with relevant statistics
consistency_stats = grouped.agg(
    model_years=('model_year', lambda x: x.nunique()),
    appearances=('name', 'count'),
    mpg_std=('mpg', 'std'),
    mpg_mean=('mpg', 'mean')
).reset_index()

# Filter consistent cars: at least 2 years and low mpg std deviation
consistent_cars = consistency_stats[
    (consistency_stats['model_years'] >= 2) &
    (consistency_stats['mpg_std'] < 1.0)
]

# Sort by appearances (descending), then average mpg (descending)
consistent_cars_sorted = consistent_cars.sort_values(
    by=['appearances', 'mpg_mean'],
    ascending=[False, False]
)

# Select and display relevant columns
print(consistent_cars_sorted[['name', 'appearances', 'mpg_mean']])

```
